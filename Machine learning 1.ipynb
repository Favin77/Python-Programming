```python
import matplotlib.image as img
from skimage.util import random_noise
char=input("Enter Input path address: ")

image=img.imread(char)

red = image[:, :, 0]
green = image[:, :, 1]
blue = image[:, :, 2]

#Red
noise_img = random_noise(red, mode='s&p',amount=0.3)#adding salt&pepper noise
noise_img = np.array(255*noise_img, dtype = 'uint8')
kernel = np.array([[-1 , -1 , -1] , [-1 , 8 , -1] ,[-1 , -1 , -1]])
red = cv2.filter2D(noise_img , -1 , kernel = kernel)

#green
noise_img = random_noise(green, mode='s&p',amount=0.3)#adding salt&pepper noise
noise_img = np.array(255*noise_img, dtype = 'uint8')
kernel = np.array([[-1 , -1 , -1] , [-1 , 8 , -1] ,[-1 , -1 , -1]])
green = cv2.filter2D(noise_img , -1 , kernel = kernel)
green=3*noise_img+green

#blue
laplacian= cv2.Laplacian(blue,cv2.CV_64F,3)
  
fig = plt.figure(figsize=(10, 7))
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(image)
#plt.axis('off')
plt.title("RGB")

fig.add_subplot(rows, columns, 2)
plt.imshow(red)
#plt.axis('off')
plt.title("RED")
  

fig.add_subplot(rows, columns, 3)
plt.imshow(green)
#plt.axis('off')
plt.title("GREEN")
  

fig.add_subplot(rows, columns, 4)
plt.imshow(blue)
#plt.axis('off')
plt.title("BLUE")
#C:\Users\favin\Desktop\College\3rdYear\2nd Semester\CV\Lab\Pepsi.jpg
```

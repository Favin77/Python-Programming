## Task 3

```python
import matplotlib.image as img
char=input("Enter Input path address: ")

image=img.imread(char)

KernelAverage2 = np.array([[1/9,1/9,1/9],[1/9,1/9,1/9],[1/9,1/9,1/9]])
KernelGauss = np.array([[0/8,1/8,0/8],[1/8,4/8,1/8],[0/8,1/8,0/8]])

image1=cv2.filter2D(image, -1, KernelAverage2)
image2=cv2.filter2D(image, -1, KernelGauss)

# create figure
fig = plt.figure(figsize=(10, 7))
  
# setting values to rows and column variables
rows = 1
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(image1)
#plt.axis('off')
plt.title("KernelAverage2")

fig.add_subplot(rows, columns, 2)
plt.imshow(image2)
#plt.axis('off')
plt.title("KernelGauss")

#C:\Users\favin\Desktop\College\3rdYear\2nd Semester\CV\Lab\Pepsi.jpg
```
